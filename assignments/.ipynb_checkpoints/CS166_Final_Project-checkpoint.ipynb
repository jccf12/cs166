## Simulating an Emergency Reponse in Manhattan

### Introduction

In this assignment I aim to simulate Emergency Response in Manhattan New York. Through the simulation, different strategies for the distribution of available ambulances among the different stations located in the city will be evaluated to determine which distribution should be chosen. The "best" distribution can be defined in more than one way, which will be discussed in this paper, but the general idea is to choose a distribution that will allow responding to emergency calls as quickly as possible.

### Representing the city as a network

The city will be represented as a network where the nodes of the network represent the neighborhoods in the city. To do so, I will base my network based on the neigborhoods of Manhattan in the following map obtained from this <a href="pinterest.com.mx/pin/814447913844402836/">source</a>:

Each node in the network will represent one of the 47 neighborhoods shown in the map above and will contain two attributes:
- **name**: the name of the Neighborhood according to the map
- **rate**: the yearly emergency rate in the given neighborhood

To determine the rate in each neighborhood, I obtained data on the number of crimes commited in each of the precincts of Manhattan. The data was obtained from the <a href="https://maps.nyc.gov/crime/">NYC government</a>. The following is the NYC Crime Map for Manhattan divided in precincts:

For the number of crimes in each precinct, I divided it uniformly among the neighborhoods within each precinct. For the precincts that only contain half a neighborhood, I assumed an even split of the number of crimes when neighborhoods are split in two. The data was inserted manually into a txt file and then imported in the Notebook that contains the code of the simulation. Here it is important to note that the number of crimes in each neighborhood is not precisely the number of emergency calls that are generated in each neighborhood. However, I assume that there is a strong correlation between the number of crimes and the emergency calls and so I will use the rate of crimes as the rate of emergency calls in the simulation.

The edges of the network represent adjacency of neighborhoods. For each two neighborhoods that are adjancent according to the neighborhoods map, there is an edge. Each of the edges has an attribute "weight" that represents the travel time between the adjacent nodes. This travel time is obtained by a google API that calcualtes the travel time between a starting and ending locations, the time is represented in minutes. Here it is important to note that the time from neighborhood A to neighborhood B is not the same as the time from B to A (due to traffic conditions and orientation of the roads in the city), for that reason the graph is a directed one.

```python
#function that gets the coordinates of given addresses using google maps api
def get_coordinates(addresses):
    coords = []
    for address in addresses:
        coords.append(str(list(gmaps.geocode(address)[0]['geometry']['location'].values()))[1:-1])
    return coords
```

```python
#obtain the distances between every pair of locations given a list of coordinates
def obtain_cost_matrix(coordinates,travel_mode = "driving",dep_time = None):
    if not dep_time:
        time_dep = datetime.datetime.now()
    else:
        ye,mo,da,ho,mi = dep_time
        t = datetime.datetime(ye,mo,da,ho,mi)
        time_dep = time.mktime(t.timetuple())

    n=len(coordinates)
    #empty matrix
    costmatrix = np.zeros([n,n])

    #obtaining travel time between each two locations
    for i in range(n):
        for j in range(n):
            if i!=j:
                start_point = coordinates[i]
                end_point = coordinates[j]
                search_result = gmaps.directions(start_point,end_point,
                                         mode=travel_mode,
                                         departure_time=time_dep
                                        ) 
                if not search_result:
                    print('GOOGLE MAPS DID NOT FIND A ROUTE BETWEEN TWO OF YOUR LOCATIONS')
                duration = search_result[0]['legs'][0]['duration']['text']
                if "hour" in duration:
                    indh = duration.index('h')
                    dur = int(duration[:indh])*60+int(duration[indh+5:-4])
                else:
                    dur = int(duration[:-4])
                costmatrix[i][j] = dur 
            else:
                costmatrix[i][j] = 0
    return costmatrix
```

```python
class Ambulance:
    def __init__(self, location, ambulance_id):
        self.location = location 
        self.time_to_assigned = 0 #time to reach the person assigned to the ambulance
        self.time_to_return = 0 #time to come back to station
        self.id = ambulance_id 
    
    def __repr__(self): #representation when one wants to print ambulance objects (used for debugging)
        return 'AMBULANCE {}'.format(self.id) 
    
    def assign_person(self, person, G): #assign a person to the Ambulance 
        self.assigned_person = person
        
        t_to_person = 0 
        dist_loc_to_person = nx.shortest_path_length(G,self.location,person.location,'weight') #get shortest path length to reach person
        dist_person_to_loc = nx.shortest_path_length(G,person.location,self.location,'weight') #get shortest path from the person back to the station
        intermediate_nodes = len(nx.shortest_path(G,self.location,person.location,'weight')) #number of intermediate nodes
        t_to_person = dist_loc_to_person + sum([np.round(np.random.normal()) for _ in range(intermediate_nodes-1)]) #adding random noise to the actual time
        t_to_return = t_to_person + dist_person_to_loc + sum([np.round(np.random.normal()) for _ in range(intermediate_nodes-1)]) #adding random noise to the actual time
        
        person.time_waited += t_to_person #the person will have waited whatever it has already waited + the time it's going to take for the ambulance to reach this person
        self.time_to_assigned = t_to_person #store information in Ambulance object attributes
        self.time_to_return = t_to_return #store information in Ambulance object attributes
```

```python
class Person:
    def __init__(self, location, person_id):
        self.location = location
        self.assigned = False
        self.time_waited = 0 #time the person has waited until she/he got the request satisfied
        
    def __gt__(self,other):
        return self.time_waited > other.time_waited #the greater person is that one who has waited for longer
                                                    #this will help for choosing people who have waited longer first
    
    def __repr__(self): #representation when one wants to print person objects (used for debugging)
        return 'P{} ASS at {}'.format(self.id, self.location) if self.assigned else 'P{} UNASS at {}'.format(self.id, self.location)
```

```python
class EmergencyResponse:
    '''
    Simulate Emergency Response in New York city
    '''

    def __init__(self,nbhoods_names, adjacent_nbhoods, rates, ambulance_stations_locations, ambulance_distribution, Gsaved = None, city_name = 'Manhattan'):
        self.nbhoods_names = nbhoods_names #neighborhood names
        self.adjacent_nbhoods = adjacent_nbhoods #adjacent neighborhood pairs
        self.rates = rates #emergency rates
        self.city_name = city_name #name of the city necessary to add to the location when calculating distances with google maps API
        self.n = len(nbhoods_names) #47 neighborhoods
        self.nbhoods_numbers = {} #dictionary to get number of neighborhood given the name
        for i in range(len(nbhoods_names)):
            self.nbhoods_numbers[nbhoods_names[i]] = i
            
        self.stations_locations = [self.nbhoods_numbers[station] for station in ambulance_stations_locations] #number of the locations of the stations 

        self.unassigned_people = [] #list of people that are still yet to be assigned
        self.satisfied_people = [] #list of people that have already been satisfied
    
        self.unassigned_ambulances = {} #dictionary of unassigned ambulances, the keys are the stations
        for station in self.stations_locations:
            self.unassigned_ambulances[station] = [] #intialize dictionary
        
        for i in range(len(ambulance_distribution)): #add an ambulance to each station according to the ambulance distribution that was input
            loc = ambulance_distribution[i] 
            self.unassigned_ambulances[station].append(Ambulance(loc,i))
            
        self.assigned_ambulances = {} #dictionary of assigned ambulances, the keys are the stations and it starts empty because no ambulanceds have been assigned
        for station in self.stations_locations:
            self.assigned_ambulances[station] = []
        
        self.Gsaved = Gsaved #store saved graph
        self.t = 0 #timesteptracker
        
    def initialize(self):
        '''
        Initialize the network based on the data for the city.
        '''
        self.graph = nx.DiGraph() 
        if self.Gsaved: #if a graph is input (we will save it later to make the intialization quicker)
            self.graph = self.Gsaved #use the input graph
        else: #otherwise initialize based on data
            nodes = [(i,
                      {'name':self.nbhoods_names[i],
                       'rate':self.rates[i]/(365), 
                       'rate_cm':self.rates[i]/max(self.rates),
                      }
                     ) for i in range(self.n)]
            self.graph.add_nodes_from(nodes)
            for pair in self.adjacent_nbhoods:
                name1, name2 = pair.split(', ')
                n1,n2 = self.nbhoods_numbers[name1], self.nbhoods_numbers[name2]
                coords = get_coordinates([name1+' '+self.city_name,name2+' '+self.city_name])
                cost_mat =  obtain_cost_matrix(coords)
                self.graph.add_edge(n1,n2, weight = cost_mat[0][1])
                self.graph.add_edge(n2,n1, weight = cost_mat[1][0])
        
        self.graph.layout = nx.spring_layout(self.graph) #spring layout
        
    def generate_requests(self, frequency = 10): #generate requests according to the emergency rate of each neighborhood
        #default frequency is every 10 minutes
        for node in list(self.graph.nodes): 
            if np.random.random() < self.graph.nodes[node]['rate']/(24*frequency): #probability of emergency call in 10 minute period according to emergency rate
                #add a person with the location at that neighborhood
                self.unassigned_people.append(Person(node,len(self.unassigned_people)+len(self.satisfied_people))) 
    
    def assign_ambulances(self):
        #as long as there are available ambulances and unassigned people, we keep assigning ambulances to people
        while self.unassigned_people and sum([len(l) for l in self.unassigned_ambulances.values()]):     
            max_wait = max(self.unassigned_people).time_waited #obtain the max wait time someone has been waiting for an ambulance
            
            next_in_line = [person for person in self.unassigned_people if person.time_waited == max_wait] 
            next_person = np.random.choice(next_in_line) #out of all those who have waited for that long choose one at random
            
            #assign next person in line
            self.unassigned_people.remove(next_person) 
            self.satisfied_people.append(next_person)
            
            stations_available = [] #check which stations have ambulances
            for station in self.unassigned_ambulances:
                if self.unassigned_ambulances[station]:
                    stations_available.append(station)
             
            #compute distance to available stations
            distances_to_stations = [nx.shortest_path_length(self.graph,next_person.location,target,'weight') for target in stations_available]
            
            #take the closest one
            closest_distance = min(distances_to_stations)
            index = distances_to_stations.index(closest_distance)
            chosen_station = stations_available[index]
            
            #assign an ambulance in the closest station to the next person
            chosen_ambulance = self.unassigned_ambulances[chosen_station].pop()
            chosen_ambulance.assign_person(next_person,self.graph)
            
            #move ambulance to assigned dictionary
            self.assigned_ambulances[chosen_station].append(chosen_ambulance)
            next_person.assigned = True #this is only relevant for printing purposes
            

    def update(self): #each time step is every 10 minutes
        
        #if it's the first update step, skip the following process since there are no people in the simulation yet
        if self.t != 0:
            for station in self.assigned_ambulances:
                for ambulance in self.assigned_ambulances[station]:
                    #update the 10 minutes that have passed for all ambulances that are currently on the process of satisfying a request
                    ambulance.time_to_assigned = max(ambulance.time_to_assigned-10,0) #advance 10 minutes, the time remaining cannot be negative
                    ambulance.time_to_return = max(ambulance.time_to_return-10,0) #advance 10 minutes, the time remaining cannot be negative
                    if ambulance.time_to_return == 0: #if the ambulance's time to return is 0, then it is done and ready to satisfy another request
                        #unassign ambulance
                        ambulance.assigned_person = None 
                        self.assigned_ambulances[station].remove(ambulance)
                        self.unassigned_ambulances[station].append(ambulance)
                        
            for person in self.unassigned_people:
                person.time_waited+=10 #add 10 minutes to the wait of all those that still are unassigned (if any)
                   
        self.generate_requests() #generate requests
        self.assign_ambulances() #satisfy requests as best as possible
        self.t+=1
        

    def observe_network(self):
        '''
        Draw the state of the network.
        '''        
        nx.draw_networkx_nodes(
                self.graph, pos=self.graph.layout, nodelist = self.stations_locations, with_labels=True, node_size = 1250, 
                edgecolors = 'black', linewidths = 6,
                    node_color='white',
                )

        nx.draw_networkx_nodes(
                self.graph, pos=self.graph.layout, with_labels=True, node_size = 1200,
                    node_color=[self.graph.nodes[i]['rate_cm'] for i in self.graph.nodes],
                    alpha=1, 
            vmin=0, vmax=1, 
            cmap = plt.cm.get_cmap('YlOrRd')
                )
        
        nx.draw_networkx_edges(self.graph,pos=self.graph.layout, arrowsize =10, node_size=1200)
        nx.draw_networkx_labels(self.graph,pos=self.graph.layout)
```

```python
#initialize first simulation to show the city as a network
nbhoods_numbers = {}
for i in range(len(nbhoods_names)):
    nbhoods_numbers[nbhoods_names[i]] = i
stations_locations = [nbhoods_numbers[station] for station in ambulance_stations]
amb_dist = [np.random.choice(stations_locations) for _ in range(10)]
sim =  EmergencyResponse(nbhoods_names,adjacent_nbhoods,crime_rates,ambulance_stations, ambulance_distribution = amb_dist)
sim.initialize()
plt.figure(figsize=(20,15))
sim.observe_network()
cbar = plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.get_cmap('YlOrRd')))
cbar.set_label('Emergency Rates (x {})'.format(max(sim.rates)),fontsize =14)
plt.show()
```

### Setting up the simulation

Three classes were built to set up the simulation: `Emergency Response`, `Ambulance` and `Person`. Each of them is explained next.

#### The `Person` class

This class is used to represent all the people that will call an ambulance. These people are in an emergency and will make requests for emergency services to come to their location. This class contains attributes for id, location, time elapsed since the request was made, whether or not the person has been assigned to an ambulance, and finally the time waited until its request was satisfied. 

#### The `Ambulance` class

This class represents each of the available ambulances in the city. The attributes are id, location, assigned person, time to its assigned person whenever it is assigned to satisfy someone's request, and time to return to its station. It also includes a method for assigning a person to the ambulance. When this method is called, the assigned person attribute is modified and each of the times to reach the assigned person and to return to the station are calculated based on the location of the person.

#### The `Emergency Response` class

This class contains all the information of the simulation. It contains the network used to represent the city, the location of the ambulances in each of the stations (which is a parameter of the simulation), the ambulance objects in the city stored in different dictionaries that correspond to whether the ambulances are assigned to a person or not, and it also contains all the people that have called for an ambulance also stored in different dictionaries that correspond to whether or not the requests of the people have been satisfied.

The methods of this class include initialization of the network based on the data for the given city, a procedure to generate requests by generating person objects based on the emergency rates for each neighborhood, a procedure to satisfy the requests of the people by the nearest ambulances based on their availability, an update method that calls the generate requests method, the assign ambulances method, and updates the appropriate time attributes of the person and ambulance objects in the simulation in each time step, and a method to observe the state of the simulation.

### Assumptions

Now, the descriptions above need more detailed explanation about how they are carried out. These explanations can be provided by laying out the assumptions used for this simulation:


- **Requests are generated uniformly throughout the day**: The way that requests are generated are based on the daily rate, calculated from the yearly rate of crimes for each neighborhood. However, the timestep that I use is 10 minutes. This means that every 10 minutes, I update the simulation by generating new requests and assigning available ambulances to those requests. To generate the requests then, I use the 10 minute emergency rate for each neighborhood (calculated from the daily rate divided by 240). Given that this rate is $r$ for a given neighborhood, I generate a person object in the given neighborhood with probability $r$. This assumes both that for every 10 minute period at most 1 emergency call can come from a given neighborhood and that every 10 minute period is equally likely to have an emergency call. 

- **For each emergency call, we always send the closest ambulance available at the moment**: I assume that whenever someone makes a call, since it is an emergency, the person deserves to receive the quickest attention possible, even if that would leave some stations without ambulances. I acknowledge there might be other strategies but I believe this one is among the most plausible ones. This is a first-come, first-serve, strategy in the sense that whoever calls first, will get be attended first and the closest ambulance will be assigned to that given person. Therefore, the way that the assign_ambulances method works is by randomly choosing a person among those with equal max waited time until the given moment to the nearest ambulance. We do this for all the waiting unassigned people unless we run out of ambulances in which case it would be necessary to wait until ambulances are unassigned after satisfying the person who called them.

- **Ambulances cannot satisfy requests until they have returned to their stations**: Whenever an ambulance is assigned, the time it will take to satisfy the request is calculated based on the shortest travel time to the location and back to the station plus some random noise that aims to simulate the time that it takes for the ambulance to reach the person within the given neighborhood and to account for uncertainties in the google map estimations. This random noise is generated by a random standarized normal distribution with mean 0 and it is added for every edge in the path to reach the desired destination and back.


### Results

To compare different distributions I ran 100 different random ways of distributing 100 ambulances among the different stations in Manhattan. The number of ambulances was chosen based on the ALS/BLS ambulances available in New York. According to <a href="https://en.wikipedia.org/wiki/New_York_City_Fire_Department_Bureau_of_EMS">Wikipedia</a>, there are 450 ALS/BLS ambulances in New York City. However, if we only consider Manhattan (excluding The Bronx, Brooklyn Central, Queens, Staten Island and Brooklyn South and Brooklyn North) there are probably only around 100 left in Manhattan (this number is somewhat arbitrary, but I belive it is reasonable assuming there are relatively a lot of people in Manhattan among the 6 divisions, so I'm assigning between 1/4 and 1/5 of the ambulance resources to Manhattan).

For each random dsitribution, 100 different simulations were run for 1440 minutes (1 day), and the time to satisfy each of the people in the 100 different simulations along the whole day in Manhattan was added to a single list from which the mean and standard deviation of the time to satisfy the requests were calculated. 

The five figures below show the distribution of the 100 ambulances among the stations for the top 5 distributions based on the expected time to satisfy the requests (the 5 that generated the smallest means). From the figures we can see how there is no clear preference we should give to any of the stations in the city. The top 5 results show how we can give preference to different stations and still get similar expected results. However, we do see how in all of our top 5 results, no stations was given more than 13 ambulances and all stations must have at least one ambulance. These results lead us to think that since no preference should be attached to any of the stations, more or less a uniform distribution should give us good results. For that reason the 5 figures that follow the first 5 were generated by running the simulation so that ambulances are evenly distributed among the stations. Since it is impossible to equally distribute the 100 ambulances among the 15 stations, for 100 different trials a random combinations to choose the five that were going to have only 6 ambulances was tried. Again the top 5 results were chosen. These results were slightly better, but just slightly. In general it seems like as long as there is somewhat evenly distributed. In fact, The lowest mean was obtained in the simulation that generated the first of the next 10 figures, which gives preference to Midtown South Station. The confidence interval in this case has an upper bound of 61.63, the second lowest out of all (only 61.47 is smaller in the 7th of the next 10 figures). Since the results are very similar for each of the shown distributions, either would be a good pick. However, the recommendation would be to use either the best one from the first 5 (the one that generated the smallest expected time to satisfy a request), or the 7th distribution, the one with smallest upper bound for the 95% confidence interval).

```python
#randomly distribute 100 ambulance
top5_results = []
for _ in range(100):
    amb_dist = [np.random.choice(stations_locations) for _ in range(100)]
    time_to_satisfy = []
    for sims in range(100):
        sim = EmergencyResponse(nbhoods_names,adjacent_nbhoods,crime_rates,ambulance_stations,amb_dist, Gsaved)
        sim.initialize()
        for timesteps in range(144):
            sim.update()
        time_to_satisfy+=[person.time_waited for person in sim.satisfied_people]
    mean = np.mean(time_to_satisfy)
    std = np.std(time_to_satisfy)
    if not top5_results:
        top5_results.append((mean,std,time_to_satisfy,amb_dist))
    elif len(top5_results)<5:
        top5_results.append((mean,std,time_to_satisfy,amb_dist))
    else:
        top5_results.remove(max(top5_results))
        top5_results.append((mean,std,time_to_satisfy,amb_dist))
```

```python
for result in top5_results:  
    mean,std,time_to_satisfy,amb_dist = result
    
    plt.figure(figsize=(16,8))
    plt.rcParams.update({'font.size': 14})
    
    plt.subplot(1,2,1)

    stations_dist_dict = {}
    for station in stations_locations:
        stations_dist_dict[nbhoods_names[station]]=0
    for station in amb_dist:
        stations_dist_dict[nbhoods_names[station]]+=1

    plt.bar([str(nbhoods_numbers[nb]) + ". " + nb for nb in stations_dist_dict.keys()],stations_dist_dict.values(), color = 'orange')
    plt.xticks(rotation=90)
    plt.title('Ambulance Distribution')
    
    plt.subplot(1,2,2)
    plt.hist(time_to_satisfy)

    plt.xlabel('The mean time to satisfy a request was {} \nwith a 95% confidence interval equal to {}'.format(
        np.round(mean,2),[np.round(mean-1.96*std,2),np.round(mean+1.96*std,2)]),fontsize=14)
    plt.title('Time to satisfy requests')
    
    plt.show()
```
